# I) Quelques transformations géométriques sur les images
On se propose ici de faire quelques transformations géométriques des images comme le retournement et la rotation.
## 1) Le retournement d'une image
Voici un programme qui permet de retourner horizontalement une image c'est à dire qui change la gauche avec la droite et réciproquement.
Lire et Exécuter le code suivant :

```python
# -*- coding: utf-8 -*-
import PIL.Image as pil
import matplotlib.pyplot as plt

# ouverture du fichier image avec la librairie pillow
img_in = pil.open("butterfly.jpg")

# affichage de l'image initiale
plt.imshow(img_in)
plt.title("butterfly.jpg")
plt.show()

# taille de l'image : L et H sont la largeur et la hauteur de l'image en pixels
L,H = img_in.size
print("definition de l'image : ", L, H)

#  création d'une image retournée horizontalement
img_out = pil.new("RGB", [L,H])
for i in range(L):
    for j in range(H):
        r,g,b = img_in.getpixel((i,j))
        img_out.putpixel((L-i-1,j), (r,g,b))

# affichage de l'image modifiée
plt.imshow(img_out)
plt.title("butterfly_inv_horizontal.jpg")
plt.show()

# sauvegarde de l'image modifiée
img_out.save("butterfly_inv_horizontal.jpg")

# fermeture du fichier créé
img_out.close()
img_in.close()
```

## 2) La rotation
On envisage ici des rotations simples à 90° vers la gauche ou la droite et uniquement pour des images carrées.
Lire et exécuter le programme suivant qui réalise une rotation à droite :

```python
# -*- coding: utf-8 -*-
import PIL.Image as pil
import matplotlib.pyplot as plt

# ouverture du fichier image avec la librairie pillow
img_in = pil.open("toucan.jpg")

# affichage de l'image initiale
plt.imshow(img_in)
plt.title("toucan.jpg")
plt.show()

# taille de l'image : L et H sont la largeur et la hauteur de l'image en pixels
L,H = img_in.size
print("definition de l'image : ", L, H)

#  création d'une image avec rotation à 90°
img_out = pil.new("RGB", [L,H])
# Parcours des colonnes i
for i in range(L):
    # Parcours des lignes j
    for j in range(H):
        r,g,b = img_in.getpixel((i,j))
        # calcul de la nouvelle position (i3,j3) du pixel (i,j)
        # 1) Translation du centre de l'image sur l'origine du repère (Oxy)
        i1 = i - L/2
        j1 = j - H/2
       
        # 2) rotation de 90° à droite : cos(90°) = 0 et sin(90°) = +1
        i2 = i1*(0) - j1*(+1)
        j2 = i1*(+1) + j1*(0)
        
        # 3) Translation du centre de l'image sur sa position initiale
        i3 = i2 + L/2
        j3 = j2 + H/2
        i3 = int(i3)
        j3 = int(j3)
        
        # placement du pixel sans déborder de l'image
        if (i3>=0 and i3<L) and (j3>=0 and j3<H):
            img_out.putpixel((i3,j3), (r,g,b))
       
# affichage de l'image modifiée
plt.imshow(img_out)
plt.title("toucan_rotation_droite.jpg")
plt.show()

# sauvegarde de l'image modifiée
img_out.save("toucan_rotation_droite.jpg")

# fermeture du fichier créé
img_out.close()
img_in.close()
```
